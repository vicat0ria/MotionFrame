```python
import cv2
import mediapipe as mp
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files
```

```python
#import mediaPose drawing utils & pose estimation model.
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
```

```python
#let users upload a video.
upload = files.upload()
video_path = list(upload.keys())[0]

#Build the video feed
cap = cv2.VideoCapture(video_path)
#if you want a more acurate model turn up detection and tracking higher^
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
  while cap.isOpened():
      ret, frame = cap.read()

      #Change colors for opencvs bgr to rgb for mediaPipe
      image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
      image.flags.writeable = False

      #detect with MP
      results = pose.process(image)

      #Color back to BGR
      image.flags.writeable = True
      image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

      #render detection
      mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

      print(results)


      cv2_imshow(image)

      #if cv2.waitKey(10) & 0xFF == ord('q'):
      if not ret:
          break

  cap.release()
  cv2.destroyAllWindows()
```
